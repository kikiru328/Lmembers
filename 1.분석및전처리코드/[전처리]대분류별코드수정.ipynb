```python
import pandas as pd
prodcat = pd.read_csv('LOTT/LOTT/상품분류.txt',encoding='cp949')
prodcat.head()
```

```python
def resort_mid_sort(df:pd.DataFrame)->pd.DataFrame:
    """
    A제휴사 대분류 1 중분류 일용삽화 ==> 대분류 9
    """
    if (df[0] =='A') and (df[1] == 1) and (df[4] == '일용잡화'):
        return 9
    else:
        return df[1]    
prodcat['대분류코드'] = prodcat.apply(lambda x: resort_mid_sort(x),axis=1)

prodcat[prodcat['중분류명']=='일용잡화']
```

```python
import pandas as pd
cust = pd.read_csv('LOTT/LOTT/고객DEMO.txt',sep = ',', encoding='cp949')
member = pd.read_csv('LOTT/LOTT/멤버십여부.txt', sep= ',', encoding='cp949')
mg = member.groupby('고객번호')[['멤버십명']].count()
mg = mg.rename(columns={'멤버십명':'멤버십가입개수'})
mg = mg.reset_index()
mg = mg.sort_values(by='멤버십가입개수',ascending=False)
cust_mem = pd.merge(cust,mg,on='고객번호',how='left')
cust_mem['멤버십가입개수'] = cust_mem['멤버십가입개수'].fillna(0).astype(int)
cust_mem = cust_mem.sort_values(by=['연령대'])
```
