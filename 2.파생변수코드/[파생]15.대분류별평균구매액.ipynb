```python
def prod_avg(df):
  #구매금액
  dfc = df[['고객번호','대분류명','구매금액']]
  dfc_g = dfc.groupby(['고객번호','대분류명']).sum()
  dfc_g = dfc_g.reset_index()
  dfc_pt = pd.pivot_table(data = dfc_g, index = '고객번호', columns = '대분류명', values = '구매금액')
  dfc_pt = dfc_pt.fillna(0).astype(int)
  
  #구매횟수
  oh = pd.get_dummies(dfc['대분류명'])
  dfc_c = pd.concat([dfc,oh],axis=1)
  dfc_c['구매횟수'] = 1
  dfc_cg = dfc_c.groupby(['고객번호','대분류명']).sum()
  dfc_cg = dfc_cg.reset_index()
  dfc_cg2 = dfc_cg.groupby('고객번호').sum()
  
  def reName(string,df):
    namelist = list(df.columns)
    for name in namelist:
      df = df.rename(columns={name:f'{name}_{string}'})
    return df

  dfc_pt_rN = reName('금액',dfc_pt)
  dfc_cg2_rN = reName('횟수',dfc_cg2)

  dfc_cg2_rN = dfc_cg2_rN.rename(columns={'구매금액_횟수':'총구매금액','구매횟수_횟수':'총구매횟수'})

  Test = pd.merge(dfc_pt_rN,dfc_cg2_rN,on='고객번호',how='left')

  sorting_cols = ['가공식품_금액','가공식품_횟수',
                 '가구/인테리어_금액','가구/인테리어_횟수',
                 '교육/문화용품_금액','교육/문화용품_횟수',
                 '기타_금액','기타_횟수',
                 '디지털/가전_금액','디지털/가전_횟수',
                 '명품_금액','명품_횟수',
                 '신선식품_금액','신선식품_횟수',
                 '의류_금액','의류_횟수',
                 '일상용품_금액','일상용품_횟수',
                 '전문스포츠/레저_금액','전문스포츠/레저_횟수',
                 '패션잡화_금액','패션잡화_횟수',
                 '총구매금액','총구매횟수']

  Test_sorted = Test[sorting_cols]
  Test_sorted = Test_sorted.astype(int)
  
  prod_list = list(df['대분류명'].unique())

  for i in prod_list:
    Test_sorted[f"{i}_구매평균액"] = Test_sorted[f'{i}_금액'] / Test_sorted[f'{i}_횟수']
  Test_sorted = Test_sorted.fillna(0)

  Test_col = list(Test_sorted.columns)
  Test_col = ['가공식품_금액','가공식품_횟수','가공식품_구매평균액',
  '가구/인테리어_금액','가구/인테리어_횟수','가구/인테리어_구매평균액',
  '교육/문화용품_금액','교육/문화용품_횟수','교육/문화용품_구매평균액',
  '기타_금액','기타_횟수','기타_구매평균액',
  '디지털/가전_금액','디지털/가전_횟수','디지털/가전_구매평균액',
  '명품_금액','명품_횟수','명품_구매평균액',
  '신선식품_금액','신선식품_횟수','신선식품_구매평균액',
  '의류_금액','의류_횟수','의류_구매평균액',
  '일상용품_금액','일상용품_횟수','일상용품_구매평균액',
  '전문스포츠/레저_금액','전문스포츠/레저_횟수','전문스포츠/레저_구매평균액',
  '패션잡화_금액','패션잡화_횟수','패션잡화_구매평균액',
  '총구매금액','총구매횟수']

  Test_sorted_rn = Test_sorted[Test_col]

  Test_sorted_rn['평균구매금액'] = Test_sorted_rn['총구매금액'] / Test_sorted_rn['총구매횟수']

  final_cols = ['가공식품_구매평균액','가구/인테리어_구매평균액','교육/문화용품_구매평균액','디지털/가전_구매평균액','명품_구매평균액','신선식품_구매평균액',
              '의류_구매평균액','일상용품_구매평균액','전문스포츠/레저_구매평균액','패션잡화_구매평균액','기타_구매평균액','평균구매금액']

  final = Test_sorted_rn[final_cols]
  return final
```

```python
H1 = prod_avg(df_1)
H2 = prod_avg(df_2)
H3 = prod_avg(df_3)
H4 = prod_avg(df_4)
```

```python
def train_df(a,b,c):

  def reName(string,df):
    namelist = list(df.columns)
    for name in namelist:
      df = df.rename(columns={name:f'{string}_{name}'})
    return df

  H1 = reName('H1',a)
  H2 = reName('H2',b)
  H3 = reName('H3',c)

  f_merge = pd.merge(H1,H2,on='고객번호',how='left')
  final_merge = pd.merge(f_merge,H3,on='고객번호',how='left')

  trainlist = ['H1','H2','H3']
  Li = list(a.columns)

  trainList = []
  for col in Li:
    for h in trainlist:
      trainList.append(f'{h}_{col}')
  final = final_merge[trainList]
  
  return final
```
