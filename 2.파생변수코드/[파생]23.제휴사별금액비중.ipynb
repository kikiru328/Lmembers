```python
def rename(df,i):
  m= df.groupby(['고객번호','제휴사']).sum('구매금액')
  pivot_m= pd.pivot_table(m, index = ['고객번호'],columns = '제휴사', values = '구매금액')
  pivot_m= pivot_m.reset_index(drop=False).rename_axis(None, axis=1)
  pivot_m= pivot_m.rename(columns={'A':f'A_{i}', 'B':f'B_{i}', 'C':f'C_{i}', 'D':f'D_{i}'})
  pivot_m = pivot_m.fillna(0)
  pivot_m['total'] = pivot_m[[f'A_{i}',f'B_{i}',f'C_{i}',f'D_{i}']].sum(axis=1)
  pivot_m[f'A_amt_r_{i}'] = pivot_m[f'A_{i}'] / pivot_m['total']
  pivot_m[f'B_amt_r_{i}'] = pivot_m[f'B_{i}'] / pivot_m['total']
  pivot_m[f'C_amt_r_{i}'] = pivot_m[f'C_{i}'] / pivot_m['total']
  pivot_m[f'D_amt_r_{i}'] = pivot_m[f'D_{i}'] / pivot_m['total']
  pivot_m = pivot_m.drop([f'A_{i}',f'B_{i}',f'C_{i}',f'D_{i}','total'],axis=1)
  return pivot_m
```
