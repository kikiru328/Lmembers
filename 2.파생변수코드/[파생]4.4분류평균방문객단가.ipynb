```python
def APV(df,a):
  countpur = df[['고객번호','영수증번호','구매일자','대분류명','구매금액']]
  countpur = countpur.groupby(['고객번호','구매일자','영수증번호','대분류명']).sum()
  countpur = countpur.reset_index()
  countpur = countpur.groupby(['구매일자','고객번호']).sum()
  countpur = countpur.reset_index()
  countpur['방문횟수'] = 1 
  countpur = countpur.groupby('고객번호').sum()
  countpur = countpur.drop('영수증번호',axis=1)
  countpur['방문객단가'] = countpur['구매금액'] / countpur['방문횟수']
  countpur = countpur.drop(['구매금액','방문횟수'],axis=1)
  cols = list(countpur.columns)
  for i in cols:
    countpur[f'{i}_{a}'] = countpur[i]
  countpur = countpur.drop(cols,axis=1)
  return countpur
```
