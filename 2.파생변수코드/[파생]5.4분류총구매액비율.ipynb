```python
df_1 = df_1[['고객번호','새분류','구매금액']]
df_2 = df_2[['고객번호','새분류','구매금액']]
df_3 = df_3[['고객번호','새분류','구매금액']]
df_4 = df_4[['고객번호','새분류','구매금액']]
```

```python
def get_pivot(df):
  test = df.groupby(['고객번호','새분류']).sum().reset_index()
  pv = pd.pivot_table(test, index='고객번호',columns='새분류',values='구매금액')
  pv = pv.reset_index()
  pv = pv.fillna(0).astype(int)
  pv['total'] = pv[['내구재','준내구재','비내구재','사치품']].sum(axis=1)
  pv = pv.reset_index(drop=True).rename_axis(None, axis=1)
  pv = pv.set_index('고객번호')
  return pv
```

```python
p1 = get_pivot(df_1)
p2 = get_pivot(df_2)
p3 = get_pivot(df_3)
p4 = get_pivot(df_4)
```

```python
p1
```

```python
def pr(df):
  Li = list(df.columns)
  for li in Li:
    df[f'{li}_amt_ratio'] = df[li] / df['total']
  df = df.drop(Li,axis=1)
  df = df.drop('total_amt_ratio',axis=1)
  df['amt_ratio_total'] = df.sum(axis=1)
  return df
```

```python
pr1 = pr(p1)
pr2 = pr(p2)
pr3 = pr(p3)
pr4 = pr(p4)
```

```python
def reName(a,df):
    cols = list(df.columns)
    for i in cols:
      df[f'{a}_{i}'] = df[i]
    A = df.drop(cols,axis=1)
    return A

H1 = reName('H1',pr1)
H2 = reName('H2',pr2)
H3 = reName('H3',pr3)
H4 = reName('H4',pr4)
```

```python
def ratio(df):
  test = df[['고객번호','새분류']]
  test = test.set_index('고객번호')
  A = pd.get_dummies(test['새분류']).reset_index()
  A = A.groupby('고객번호').sum().astype(int)
  Li = list(A.columns)
  A['total'] = A.sum(axis=1).astype(int)
  for i in Li:
    A[f"{i}_r"] = A[i] / A['total']
  B = A.drop(['내구재','사치품','비내구재','준내구재','total'],axis=1)
  B['total_r'] = B.sum(axis=1)
  return B
```
