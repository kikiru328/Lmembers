```python
def comp(compdf):
  compdf = compdf.groupby(['고객번호','제휴사']).sum()
  compdf = compdf.reset_index()
  compdf = compdf.drop(['연도','월'],axis=1)
  return compdf

def finalcomp(productdf,compdf):
  productdf = productdf.groupby(['고객번호','제휴사'])[['구매횟수']].sum()
  productdf = productdf.reset_index()
  Test = pd.merge(productdf,compdf,on=['고객번호','제휴사'],how='outer')
  Test = Test.fillna(0)
  Test = Test.sort_values(by='경쟁사이용횟수',ascending=False)
  Test[['고객번호','구매횟수', '경쟁사이용횟수', 'A01', 'A02', 'B01', 'B02', 'C01','C02', 'C03', 'D01', 'D02']] = Test[['고객번호','구매횟수', '경쟁사이용횟수', 'A01', 'A02', 'B01', 'B02', 'C01','C02', 'C03', 'D01', 'D02']].astype(int)
  Test = Test.groupby('고객번호').agg({'제휴사':lambda x: ','.join(x),
                                '구매횟수':'sum',
                                '경쟁사이용횟수':'sum',
                                'A01':'sum',
                                'A02':'sum',
                                'B01':'sum',
                                'B02':'sum',
                                'C01':'sum',
                                'C02':'sum',
                                'D01':'sum',
                                'D02':'sum'})
  return Test
```
